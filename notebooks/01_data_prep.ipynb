```python

%pip install pandas numpy scikit-learn matplotlib seaborn jupyter

import pandas as pd
import numpy as np
from itertools import combinations
import sqlite3
import os

DATA_PATH  = '../data/'
MODEL_PATH = '../model/'
os.makedirs(MODEL_PATH, exist_ok=True)
print("Setup done!")
```

```python
orders       = pd.read_csv(f'{DATA_PATH}orders.csv')
op_train     = pd.read_csv(f'{DATA_PATH}order_products__train.csv')
op_prior     = pd.read_csv(f'{DATA_PATH}order_products__prior.csv')
products     = pd.read_csv(f'{DATA_PATH}products.csv')
aisles       = pd.read_csv(f'{DATA_PATH}aisles.csv')
departments  = pd.read_csv(f'{DATA_PATH}departments.csv')

products = products.merge(aisles, on='aisle_id')
products = products.merge(departments, on='department_id')

print(f"Products loaded: {len(products)}")
print(products[['product_id','product_name','aisle','department']].head(10))
```

```python
conn = sqlite3.connect(f'{DATA_PATH}instacart.db')
products.to_sql('products', conn, if_exists='replace', index=False)
orders.to_sql('orders',   conn, if_exists='replace', index=False)
conn.close()
print("Saved to instacart.db!")
```

```python
all_op = pd.concat([op_prior, op_train])
order_groups = all_op.groupby('order_id')['product_id'].apply(list)

print("Building positive pairs... (takes 2-3 mins)")
positive_pairs = []
for prods in order_groups:
    if len(prods) >= 2:
        for p1, p2 in combinations(prods[:6], 2):
            positive_pairs.append((p1, p2, 1))

pos_df = pd.DataFrame(positive_pairs, columns=['product_1','product_2','label'])
pos_df = pos_df.drop_duplicates().sample(n=100000, random_state=42)
print(f"Positive pairs: {len(pos_df)}")
```

```python
print("Building negative pairs...")
all_ids = products['product_id'].tolist()
neg = set()
while len(neg) < 100000:
    p1, p2 = np.random.choice(all_ids, 2, replace=False)
    neg.add((int(p1), int(p2), 0))

neg_df = pd.DataFrame(list(neg), columns=['product_1','product_2','label'])
print(f"Negative pairs: {len(neg_df)}")
```

```python
pairs_df = pd.concat([pos_df, neg_df]).sample(frac=1, random_state=42).reset_index(drop=True)
pairs_df.to_csv(f'{DATA_PATH}pairs.csv', index=False)
print(f"Total pairs saved: {len(pairs_df)}")
print(pairs_df['label'].value_counts())
```
